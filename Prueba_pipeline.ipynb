He aquí un ejemplo de clasificación de secuencias utilizando un modelo para determinar si dos secuencias son paráfrasis la una de la otra. El proceso es el siguiente:

- Instanciar un tokenizador y un modelo a partir del nombre del punto de control. El modelo se identifica como un modelo BERT y se carga con los pesos almacenados en el punto de control.
- Se construye una secuencia a partir de las dos frases, con los separadores correctos específicos del modelo, los identificadores de tipo de token y las máscaras de atención (que creará automáticamente el tokenizador).
- Pase esta secuencia por el modelo para que se clasifique en una de las dos clases disponibles: 0 (no es una paráfrasis) y 1 (es una paráfrasis).
- Calcule el softmax del resultado para obtener las probabilidades sobre las clases.
- Imprima los resultados.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased-finetuned-mrpc")
model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased-finetuned-mrpc")

classes = ["not paraphrase", "is paraphrase"]

sequence_0 = "The company HuggingFace is based in New York City"
sequence_1 = "Apples are especially bad for your health"
sequence_2 = "HuggingFace's headquarters are situated in Manhattan"

# The tokenizer will automatically add any model specific separators (i.e. <CLS> and <SEP>) and tokens to
# the sequence, as well as compute the attention masks.
paraphrase = tokenizer(sequence_0, sequence_2, return_tensors="pt")
not_paraphrase = tokenizer(sequence_0, sequence_1, return_tensors="pt")

paraphrase_classification_logits = model(**paraphrase).logits
not_paraphrase_classification_logits = model(**not_paraphrase).logits

paraphrase_results = torch.softmax(paraphrase_classification_logits, dim=1).tolist()[0]
not_paraphrase_results = torch.softmax(not_paraphrase_classification_logits, dim=1).tolist()[0]

# Should be paraphrase
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(paraphrase_results[i] * 100))}%")
```

```python
# Should not be paraphrase
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(not_paraphrase_results[i] * 100))}%")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
import tensorflow as tf

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased-finetuned-mrpc")
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-cased-finetuned-mrpc")

classes = ["not paraphrase", "is paraphrase"]

sequence_0 = "The company HuggingFace is based in New York City"
sequence_1 = "Apples are especially bad for your health"
sequence_2 = "HuggingFace's headquarters are situated in Manhattan"

# The tokenizer will automatically add any model specific separators (i.e. <CLS> and <SEP>) and tokens to
# the sequence, as well as compute the attention masks.
paraphrase = tokenizer(sequence_0, sequence_2, return_tensors="tf")
not_paraphrase = tokenizer(sequence_0, sequence_1, return_tensors="tf")

paraphrase_classification_logits = model(paraphrase).logits
not_paraphrase_classification_logits = model(not_paraphrase).logits

paraphrase_results = tf.nn.softmax(paraphrase_classification_logits, axis=1).numpy()[0]
not_paraphrase_results = tf.nn.softmax(not_paraphrase_classification_logits, axis=1).numpy()[0]

# Should be paraphrase
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(paraphrase_results[i] * 100))}%")
```

```python
# Should not be paraphrase
for i in range(len(classes)):
    print(f"{classes[i]}: {int(round(not_paraphrase_results[i] * 100))}%")
```

He aquí un ejemplo de respuesta a preguntas utilizando un modelo y un tokenizador. El proceso es el siguiente:

- Instanciar un tokenizador y un modelo a partir del nombre del punto de control. El modelo se identifica como un modelo BERT y se carga con los pesos almacenados en el punto de control.
- Se define un texto y unas preguntas.
- Se itera sobre las preguntas y se construye una secuencia a partir del texto y la pregunta actual, con los separadores correctos específicos del modelo, los identificadores de tipo de token y las máscaras de atención.
- Pase esta secuencia por el modelo. Se obtiene un rango de puntuaciones de toda la secuencia de tokens (pregunta y texto), tanto para la posición inicial como para la final.
- Calcule el softmax del resultado para obtener probabilidades sobre los tokens.
- Obtenga los tokens a partir de los valores de inicio y fin identificados y conviértalos en una cadena.
- Imprima los resultados.


```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
model = AutoModelForQuestionAnswering.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")

text = r"""
🤗 Transformers (formerly known as pytorch-transformers and pytorch-pretrained-bert) provides general-purpose
architectures (BERT, GPT-2, RoBERTa, XLM, DistilBert, XLNet…) for Natural Language Understanding (NLU) and Natural
Language Generation (NLG) with over 32+ pretrained models in 100+ languages and deep interoperability between
TensorFlow 2.0 and PyTorch.
"""

questions = [
    "How many pretrained models are available in 🤗 Transformers?",
    "What does 🤗 Transformers provide?",
    "🤗 Transformers provides interoperability between which frameworks?",
]

for question in questions:
    inputs = tokenizer(question, text, add_special_tokens=True, return_tensors="pt")
    input_ids = inputs["input_ids"].tolist()[0]

    outputs = model(**inputs)
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits

    # Get the most likely beginning of answer with the argmax of the score
    answer_start = torch.argmax(answer_start_scores)
    # Get the most likely end of answer with the argmax of the score
    answer_end = torch.argmax(answer_end_scores) + 1

    answer = tokenizer.convert_tokens_to_string(
        tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end])
    )

    print(f"Question: {question}")
    print(f"Answer: {answer}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForQuestionAnswering
import tensorflow as tf

tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
model = TFAutoModelForQuestionAnswering.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")

text = r"""
🤗 Transformers (formerly known as pytorch-transformers and pytorch-pretrained-bert) provides general-purpose
architectures (BERT, GPT-2, RoBERTa, XLM, DistilBert, XLNet…) for Natural Language Understanding (NLU) and Natural
Language Generation (NLG) with over 32+ pretrained models in 100+ languages and deep interoperability between
TensorFlow 2.0 and PyTorch.
"""

questions = [
    "How many pretrained models are available in 🤗 Transformers?",
    "What does 🤗 Transformers provide?",
    "🤗 Transformers provides interoperability between which frameworks?",
]

for question in questions:
    inputs = tokenizer(question, text, add_special_tokens=True, return_tensors="tf")
    input_ids = inputs["input_ids"].numpy()[0]

    outputs = model(inputs)
    answer_start_scores = outputs.start_logits
    answer_end_scores = outputs.end_logits

    # Get the most likely beginning of answer with the argmax of the score
    answer_start = tf.argmax(answer_start_scores, axis=1).numpy()[0]
    # Get the most likely end of answer with the argmax of the score
    answer_end = tf.argmax(answer_end_scores, axis=1).numpy()[0] + 1

    answer = tokenizer.convert_tokens_to_string(
        tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end])
    )

    print(f"Question: {question}")
    print(f"Answer: {answer}")
```

**Modelización lingüística**

El modelado del lenguaje es la tarea de ajustar un modelo a un corpus, que puede ser específico de un dominio. Todos los modelos populares basados en transformadores se entrenan utilizando una variante del modelado lingüístico, por ejemplo, BERT con modelado lingüístico enmascarado, GPT-2 con modelado lingüístico causal.

El modelado del lenguaje también puede ser útil fuera del preentrenamiento, por ejemplo para cambiar la distribución del modelo para que sea específica de un dominio: utilizando un modelo de lenguaje entrenado en un corpus muy grande y ajustándolo después a un conjunto de datos de noticias o a artículos científicos, por ejemplo LysandreJik/arxiv-nlp.


**Modelado de lenguaje enmascarado**

El modelado del lenguaje enmascarado consiste en enmascarar las señales de una secuencia con una señal de enmascaramiento y pedir al modelo que rellene la máscara con una señal adecuada. De este modo, el modelo puede prestar atención tanto al contexto derecho (fichas a la derecha de la máscara) como al contexto izquierdo (fichas a la izquierda de la máscara). Un entrenamiento de este tipo crea una base sólida para tareas posteriores que requieren un contexto bidireccional, como SQuAD (respuesta a preguntas, véase Lewis, Lui, Goyal et al., parte 4.2). Si desea afinar un modelo en una tarea de modelado lingüístico enmascarado, puede aprovechar el script run_mlm.py.

He aquí un ejemplo de uso de pipelines para sustituir una máscara de una secuencia:

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
```

He aquí un ejemplo de modelado de lenguaje enmascarado utilizando un modelo y un tokenizador. El proceso es el siguiente:

- Instanciar un tokenizador y un modelo a partir del nombre del punto de control. El modelo se identifica como un modelo DistilBERT y se carga con los pesos almacenados en el punto de control.
- Definir una secuencia con un token enmascarado, colocando el tokenizer.mask_token en lugar de una palabra.
- Codifica esa secuencia en una lista de IDs y encuentra la posición del token enmascarado en esa lista.
- Recuperar las predicciones en el índice del token enmascarado: este tensor tiene el mismo tamaño que el vocabulario, y los valores son las puntuaciones atribuidas a cada token. El modelo otorga una puntuación más alta a las palabras que considera probables en ese contexto.
- Recupere los 5 tokens principales utilizando los métodos PyTorch topk o TensorFlow top_k.
- Sustituye la máscara de tokens por los tokens e imprime los resultados

```python
from transformers import AutoModelForMaskedLM, AutoTokenizer
import torch

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased")
model = AutoModelForMaskedLM.from_pretrained("distilbert-base-cased")

sequence = (
    "Distilled models are smaller than the models they mimic. Using them instead of the large "
    f"versions would help {tokenizer.mask_token} our carbon footprint."
)

inputs = tokenizer(sequence, return_tensors="pt")
mask_token_index = torch.where(inputs["input_ids"] == tokenizer.mask_token_id)[1]

token_logits = model(**inputs).logits
mask_token_logits = token_logits[0, mask_token_index, :]

top_5_tokens = torch.topk(mask_token_logits, 5, dim=1).indices[0].tolist()

for token in top_5_tokens:
    print(sequence.replace(tokenizer.mask_token, tokenizer.decode([token])))
```

```python
from transformers import TFAutoModelForMaskedLM, AutoTokenizer
import tensorflow as tf

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-cased")
model = TFAutoModelForMaskedLM.from_pretrained("distilbert-base-cased")

sequence = (
    "Distilled models are smaller than the models they mimic. Using them instead of the large "
    f"versions would help {tokenizer.mask_token} our carbon footprint."
)

inputs = tokenizer(sequence, return_tensors="tf")
mask_token_index = tf.where(inputs["input_ids"] == tokenizer.mask_token_id)[0, 1]

token_logits = model(**inputs).logits
mask_token_logits = token_logits[0, mask_token_index, :]

top_5_tokens = tf.math.top_k(mask_token_logits, 5).indices.numpy()

for token in top_5_tokens:
    print(sequence.replace(tokenizer.mask_token, tokenizer.decode([token])))
```

**Modelado causal del lenguaje**

La modelización causal del lenguaje consiste en predecir el token que sigue a una secuencia de tokens. En esta situación, el modelo sólo tiene en cuenta el contexto izquierdo (los tokens situados a la izquierda de la máscara). Este tipo de entrenamiento es especialmente interesante para las tareas de generación. Si desea afinar un modelo en una tarea de modelado causal del lenguaje, puede aprovechar el script run_clm.py.

Normalmente, el siguiente token se predice mediante el muestreo de los logits del último estado oculto que el modelo produce a partir de la secuencia de entrada.

He aquí un ejemplo de uso del tokenizador y el modelo y de aprovechamiento del método top_k_top_p_filtering() para muestrear el siguiente token a partir de una secuencia de tokens de entrada.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, top_k_top_p_filtering
import torch
from torch import nn

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

sequence = f"Hugging Face is based in DUMBO, New York City, and"

inputs = tokenizer(sequence, return_tensors="pt")
input_ids = inputs["input_ids"]

# get logits of last hidden state
next_token_logits = model(**inputs).logits[:, -1, :]

# filter
filtered_next_token_logits = top_k_top_p_filtering(next_token_logits, top_k=50, top_p=1.0)

# sample
probs = nn.functional.softmax(filtered_next_token_logits, dim=-1)
next_token = torch.multinomial(probs, num_samples=1)

generated = torch.cat([input_ids, next_token], dim=-1)

resulting_string = tokenizer.decode(generated.tolist()[0])
print(resulting_string)
```

A continuación se muestra un ejemplo de uso del tokenizador y el modelo y de aprovechamiento del método tf_top_k_top_p_filtering() para muestrear el siguiente token tras una secuencia de tokens de entrada.

```python
from transformers import TFAutoModelForCausalLM, AutoTokenizer, tf_top_k_top_p_filtering
import tensorflow as tf

tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = TFAutoModelForCausalLM.from_pretrained("gpt2")

sequence = f"Hugging Face is based in DUMBO, New York City, and"

inputs = tokenizer(sequence, return_tensors="tf")
input_ids = inputs["input_ids"]

# get logits of last hidden state
next_token_logits = model(**inputs).logits[:, -1, :]

# filter
filtered_next_token_logits = tf_top_k_top_p_filtering(next_token_logits, top_k=50, top_p=1.0)

# sample
next_token = tf.random.categorical(filtered_next_token_logits, dtype=tf.int32, num_samples=1)

generated = tf.concat([input_ids, next_token], axis=1)

resulting_string = tokenizer.decode(generated.numpy().tolist()[0])
print(resulting_string)
```

Esto produce un token (esperemos) coherente que sigue a la secuencia original, que en nuestro caso es la palabra is o features.

**Reconocimiento de Entidades Nombradas**

El reconocimiento de entidades con nombre (NER) es la tarea de clasificar los tokens según una clase, por ejemplo, identificando un token como una persona, una organización o un lugar. Un ejemplo de conjunto de datos de reconocimiento de entidades con nombre es el conjunto de datos CoNLL-2003, que se basa enteramente en esa tarea. Si desea ajustar un modelo en una tarea NER, puede utilizar el script run_ner.py.

A continuación se muestra un ejemplo de uso de pipelines para el reconocimiento de entidades con nombre, en concreto, tratando de identificar los tokens como pertenecientes a una de las 9 clases:

- O, Fuera de una entidad con nombre
- B-MIS, Comienzo de una entidad miscelánea justo después de otra entidad miscelánea
- I-MIS, entidad miscelánea
- B-PER, Comienzo del nombre de una persona justo después del nombre de otra persona
- I-PER, Nombre de persona
- B-ORG, Comienzo de una organización justo después de otra organización
- I-ORG, Organización
- B-LOC, Inicio de un lugar justo después de otro lugar
- I-LOC, Ubicación

Aprovecha un modelo perfeccionado sobre CoNLL-2003, perfeccionado por @stefan-it de dbmdz.

```python
from transformers import pipeline

ner_pipe = pipeline("ner")

sequence = """Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO,
therefore very close to the Manhattan Bridge which is visible from the window."""
```

Esto produce una lista de todas las palabras que han sido identificadas como una de las entidades de las 9 clases definidas anteriormente. Estos son los resultados esperados:

```python
for entity in ner_pipe(sequence):
    print(entity)
```

Observe cómo las fichas de la secuencia "Cara de abrazo" se han identificado como una organización, y "Ciudad de Nueva York",
"DUMBO" y "Manhattan Bridge" se han identificado como lugares.

He aquí un ejemplo de reconocimiento de entidades con nombre, utilizando un modelo y un tokenizador. El proceso es el siguiente

1. Instanciar un tokenizador y un modelo a partir del nombre del punto de control. El modelo se identifica como un modelo BERT y lo carga
   con los pesos almacenados en el punto de control.
2. 2. Definir una secuencia con entidades conocidas, como "Hugging Face" como organización y "New York City" como ubicación.
3. 3. Dividir las palabras en tokens para poder asignarlas a predicciones. Para ello, primero codificamos y descodificamos
   codificando y descodificando la secuencia, de modo que nos quede una cadena que contenga los tokens especiales.
4. 4. Codificamos la secuencia en IDs (los tokens especiales se añaden automáticamente).
5. Recuperar las predicciones pasando la entrada al modelo y obteniendo la primera salida. El resultado es una
   sobre las 9 clases posibles para cada token. Tomamos el argmax para obtener la clase más probable para cada token.
   para cada símbolo.
6. Juntamos cada ficha con su predicción y la imprimimos.

```python
from transformers import AutoModelForTokenClassification, AutoTokenizer
import torch

model = AutoModelForTokenClassification.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = (
    "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, "
    "therefore very close to the Manhattan Bridge."
)

inputs = tokenizer(sequence, return_tensors="pt")
tokens = inputs.tokens()

outputs = model(**inputs).logits
predictions = torch.argmax(outputs, dim=2)
```

```python
from transformers import TFAutoModelForTokenClassification, AutoTokenizer
import tensorflow as tf

model = TFAutoModelForTokenClassification.from_pretrained("dbmdz/bert-large-cased-finetuned-conll03-english")
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = (
    "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, "
    "therefore very close to the Manhattan Bridge."
)

inputs = tokenizer(sequence, return_tensors="tf")
tokens = inputs.tokens()

outputs = model(**inputs)[0]
predictions = tf.argmax(outputs, axis=2)
```

Esto produce una lista de cada token asignada a su predicción correspondiente. A diferencia de la línea de producción, aquí cada token tiene una predicción, ya que no hemos eliminado la clase "0", lo que significa que no se ha encontrado ninguna entidad concreta en ese token.

En el ejemplo anterior, predictions es un número entero que corresponde a la clase predicha. Podemos utilizar la propiedad model.config.id2label para recuperar el nombre de la clase correspondiente al número de clase, como se ilustra a continuación:

```python
for token, prediction in zip(tokens, predictions[0].numpy()):
    print((token, model.config.id2label[prediction]))
```

**Clasificación de audio **

La clasificación de audio asigna una clase a una señal de audio. El conjunto de datos Keyword Spotting de SUPERB es un ejemplo de conjunto de datos que puede utilizarse para ajustar la clasificación de audio. Este conjunto de datos contiene diez clases de palabras clave para la clasificación. Si desea ajustar un modelo de clasificación de audio, consulte el script run_audio_classification.py o esta guía práctica.

Los siguientes ejemplos demuestran cómo utilizar un pipeline() y un modelo y tokenizer para la inferencia de clasificación de audio:

```python
from transformers import pipeline
from datasets import load_dataset
import torch

torch.manual_seed(42)
dataset = load_dataset("hf-internal-testing/librispeech_asr_demo", "clean", split="validation")
dataset = dataset.sort("id")
audio_file = dataset[0]["audio"]["path"]

audio_classifier = pipeline(
    task="audio-classification", model="ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition"
)
predictions = audio_classifier(audio_file)
predictions = [{"score": round(pred["score"], 4), "label": pred["label"]} for pred in predictions]
predictions
```

El proceso general para utilizar un modelo y un extractor de características para la clasificación de audio es:

- Instanciar un extractor de características y un modelo a partir del nombre del punto de control.
- Procesar la señal de audio a clasificar con un extractor de características.
- Pasar la entrada a través del modelo y tomar el argmax para recuperar la clase más probable.
- Convertir el id de clase en un nombre de clase con id2label para devolver un resultado interpretable.

```python
from transformers import AutoFeatureExtractor, AutoModelForAudioClassification
from datasets import load_dataset
import torch

dataset = load_dataset("hf-internal-testing/librispeech_asr_demo", "clean", split="validation")
dataset = dataset.sort("id")
sampling_rate = dataset.features["audio"].sampling_rate

feature_extractor = AutoFeatureExtractor.from_pretrained("superb/wav2vec2-base-superb-ks")
model = AutoModelForAudioClassification.from_pretrained("superb/wav2vec2-base-superb-ks")

inputs = feature_extractor(dataset[0]["audio"]["array"], sampling_rate=sampling_rate, return_tensors="pt")

with torch.no_grad():
    logits = model(**inputs).logits

predicted_class_ids = torch.argmax(logits, dim=-1).item()
predicted_label = model.config.id2label[predicted_class_ids]
predicted_label
```

El proceso general para utilizar un modelo y un procesador para el reconocimiento automático del habla es el siguiente:

- Instanciar un procesador (que reagrupa un extractor de características para el tratamiento de la entrada y un tokenizador para la descodificación) y un modelo a partir del nombre del punto de control.
- Procesar la señal de audio y el texto con un procesador.
- Pasa la entrada por el modelo y toma el argmax para recuperar el texto predicho.

```python
from transformers import AutoProcessor, AutoModelForCTC
from datasets import load_dataset
import torch

dataset = load_dataset("hf-internal-testing/librispeech_asr_demo", "clean", split="validation")
dataset = dataset.sort("id")
sampling_rate = dataset.features["audio"].sampling_rate

processor = AutoProcessor.from_pretrained("facebook/wav2vec2-base-960h")
model = AutoModelForCTC.from_pretrained("facebook/wav2vec2-base-960h")

inputs = processor(dataset[0]["audio"]["array"], sampling_rate=sampling_rate, return_tensors="pt")
with torch.no_grad():
    logits = model(**inputs).logits
predicted_ids = torch.argmax(logits, dim=-1)

transcription = processor.batch_decode(predicted_ids)
transcription[0]
```
