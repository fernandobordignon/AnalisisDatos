**Decisiones**

La instrucción **if** permite manejar el **control de flujo** dentro de un programa. A partir de una evaluación de una expresión se puede indicar que se haga algo o no.

```python
# Vemos la aplicación de la instrucción if, en un primer caso la expresión "horas_extra>0" se valora como verdadera 
# y en el segundo como falsa

horas_extra = 15

if(horas_extra > 0):
    total = horas_extra*24.4
    print("El adicional a cobrar por horas extra es "+str(total)+" pesos")
else:
    print("No hay adicional por que no se registrar horas extras")

    
horas_extra = 0

if(horas_extra > 0):
    total = horas_extra*24.4
    print("El adicional a cobrar por horas extra es "+str(total)+" pesos")
else:
    print("No hay adicional por que no se registrar horas extras")
```

**Ciclos** 

Son construcciones que nos permiten iterar una cantidad definida de veces por un subconjunto de instrucciones

```python
contador = 0
while contador < 5:
    print("“El valor almacenado en la variable contador es ",str(contador))
    contador = contador + 1
```

**Funciones**

Son **pequeños programas que el programador puede diseñar e invocarlos** en la ejecución de un programa principal. Las funciones permiten extender las posibilidades del lenguaje como así también hacer más ordenada, eficiente y legible la programación.

```python
# Ejemplo de función. Devuelve p si un número consultado es par o i si es impar

def par_impar(x):
    if x % 2 == 0:   
        return("p")
    else:
        return("i")

# Ahora invocaremos a la función con distintos números que serán pasados como parámetros

print(par_impar(3443))
print(par_impar(-344))
print(par_impar(-7))
```

Estructura de datos **lista** (list)

Una lista es una estructura de datos que nos permiten almacenar un grupo datos de cualquier tipo de valor. Ejmplos de listas son:

```python
lista1 =["bananas", "peras", "limones", "frutillas", "cerezas"]

print(lista1)
print(len(lista1)) # len devuelve el largo de una lista
print(lista1[1])  # muestra el elemento i de una lista. Los elementos de una lista empiezan de cero
print("Debo comprar ",lista1[0]," y ",lista1[4])
```

```python
lista2 = [-34, 33, -43.7, 122, 341, -500, 0, 180]

# Es posible crer sublistas a partir de una lista (slices).

print(lista2[0:2]) # devuelve la lista [-34,33]
print(lista2[4:])  # devuelve desde el elemento 4 todos los restantes 
print(lista2[4:7])# devuelve [341, -500, 0]
```

```python
# Un elemento de una lista se modifica de la siguiente manera

lista2[2] = 222

# Añade un elemento al final de la lista

lista2.append(777)

# Inserta un elemento en la lista en la posición 2

lista2.insert(6, -33)

lista2
```

```python
# Borra un elemento de la lista

lista2.remove(0) # remueve los elementos de la lista cuyo valor es cero

lista2
```

```python
del lista2[2] # borra el elemento 2 de la  lista

lista2

```
